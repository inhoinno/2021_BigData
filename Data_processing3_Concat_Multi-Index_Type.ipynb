## 1. Load Data with Multi-level indexing
pseudo code : 
* for FILE in range(2019_1, 2020_12)
    * FILE = './1920_processed/{year}_{month}_processed.csv'.format(year=Y, month=M) 
    * olddata = newdata
    * newdata = pd.read_csv(FILE,encoding='utf-8')
    * newdata['year'] = Y 
    * newdata['month'] = M
    * newdata = newdata.set_index(['year','month', 'A. 도시명'])
    * newdata.index.names=['year','month','city']
    * newdata = newdata.sort_index()
    * pd.concat([olddata,newdata], axis=0)

```python
Y=2019
M=1
olddata = pd.DataFrame()
newdata = pd.DataFrame()
```

```python
# File
while(Y<2021):
    FILE = './1920_processed/{year}_{month}_processed.csv'.format(year=Y, month=M) 
    
    # Read CSV
    #if(Y==2019 & M ==1):
    newdata = pd.read_csv(FILE,encoding='utf-8')
    newdata = newdata.drop(columns = newdata.columns[0])
    #else:
        #olddata = newdata
        #newdata = pd.read_csv(FILE,encoding='utf-8')
        #newdata = data.drop(columns = data.columns[0])
        
    #MultiIndexing 
    newdata['year'] = Y
    newdata['month'] = M
    newdata = newdata.set_index(['year','month','A. 도시명'])
    newdata.index.names=['year','month','city']
    newdata = newdata.sort_index()
    
    if (Y==2019 & M ==1):
        olddata = newdata
    else :
        olddata = pd.concat([olddata,newdata], axis = 0)
    print('{0}_{1} fin'.format(Y,M))
    # increment
    if(M + 1 >= 13): # M=12 Next year
        M = 1
        Y += 1
    else :     
        M += 1 
```
