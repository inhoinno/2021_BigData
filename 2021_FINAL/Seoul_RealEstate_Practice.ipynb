```python
data['시군구'].values[0].split()
```

```python
data['시'] = data.시군구.str.split().str[0]
data['군'] = data.시군구.str.split().str[1]
data['구'] = data.시군구.str.split().str[2]
```

```python
data['계약년월'] = data['계약년월'].astype(str)
data['계약 년'] = data.계약년월.str[:4]
data['계약 월'] = data.계약년월.str[4:]
```

```python
data = data.drop(columns=['해제사유발생일','시군구','계약일'])
```

```python
data= data[['시', '군', '구', '계약 년', '계약 월', '본번', '부번', '단지명', '전용면적(㎡)','층', '건축년도', '거래금액(만원)', '도로명']]
data= data.sort_values(by=['계약 월'],ascending=True)
data
```

```python
data=data.reset_index(drop=True)
```

```python
data['전용면적(㎡)'].astype('float').hist(bins = 100, density=True)
```

```python
plt.hist(data['전용면적(㎡)'].astype('float'), bins = 100, density=True)
plt.yscale('log')
```

```python
data['전용면적(㎡)'] = data['전용면적(㎡)'].astype(float)
data['거래금액(만원)'] = data['거래금액(만원)'].str.replace(",","").astype(int)
```

```python
def evaluating(x):
    if x <50 :
        return 'A'
    elif x<=100 :
        return 'B'
    elif x<=150 :
        return 'C'
    else :
        return 'D'
# A <= 50 15평
# B <=100 30평
# C <=150 45평
# D >150  45평 이상
```

```python
data['area'] = data['전용면적(㎡)'].apply(evaluating)
```

```python
data['price_per_area'] = data['거래금액(만원)']/data['전용면적(㎡)']
```

```python
data['area'] 
data['price_per_area']
```

```python
df = data[['구','area','거래금액(만원)','price_per_area']].groupby(by=['구','area']).mean()
```
