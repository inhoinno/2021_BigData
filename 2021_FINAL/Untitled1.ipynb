### 4. 결론
2006년부터 2021년까지 서울시 전체와 각 자치구별 집값에 대해서 살펴보았다.
2. CoarseGrain : 서울시 전체 집 값 추이에서는 실거래가 기준 ABCD(평수 별)로 각각 A:15평이하, B:15평이상 30평이하, C:30평 이상 45평 이하, D:45평이상 으로 나누어 집값 동향을 살펴보았다. 결과적으로 2006년 대비 약 3배, 2015년 대비 2배라는 결과를 얻을 수 있었다.<br>
서울시 1평당 집값 추이에서는 2006년 A,B,C,D순으로 평수가 더 높을 수록 평 단가는 떨어지는 현상을 보였고, 이는 조금 수정되어 2021년 B,A,C,D순으로 1평당 집값이 변경되었다. 다른 말로 하자면 15평이상 30평 이하의 집이 15평 이하보다 수요가 많음을 의미하고, 1인 가구와 젊은 신혼 부부의 수요가 합쳐진 결과로 해석된다.<br>
<br>
그 다음으로는, 문재인 정부가 지난 25차례 발표한 부동산 대책 시기별 집값 동향을 살펴보았다. 전체적으로 보았을 때, 25차례의 횟수와 예산의 규모에 무색하게 집값은 떨어지지 않았으며, 잡히지 않았다. 다만 2021년 강남구와 마포구 등에서 A형 집(15평 이하)의 매매가가 다소 감소하는 추세는 확인되었다. <br>
<br>
또 KOSPI와 함께본 서울시 전체 집값은 매우 유사하게 전개되었으며, 특히 D형(45평 이상)의 집값이 KOSPI의 상승 시기 및 횡보와 유사하게 전개되었다. 특이점은 KOSPI는 등락이 있었으나, 집값은 2008년 세계 경제 대공황 이후를 제외하고 2014-2015년 부터는 계속해서 상승한 것을 볼수 있었다.<br>
<br>
3. FineGrain : 서울특별시 자치구 별 집값 추이에서는 각 자치구별 집값에 대한 동향을 볼 수 있었다. 2021년 기준 강남구, 서초구, 송파구, 용산구, 성동구 등이 높은 아파트 매매가를 기록하였다. 반면 강북구, 노원구, 0금천구, 관악구, 도봉구 등은 서울에서 대부분 집값이 그나마 싼 지역으로 확인되었다.<br>
<br>
같은 결과를 2006년부터 보았을 때, 가장 높은 집값을 기록한 자치구는 "강남구", 가장 낮은 집값을 기록한 자치구는 "강북구"로 확인되었다. 다만 서술의 편의를 위해 1개씩 자치구를 골라 그 동향을 설명하였을 뿐, 오차는 존재한다.<br>
<br>
서울시 자치구별 집값 변화 추이를 보았을 때, y값을 공유하지 않는 그래프 기준 전부 상승폭이 일정하게 큰 것을 알 수 있었다. 그중 "노도강금관구"의 경우, "강남3구+마용성"보다 집값 상승률이 컸다. 혹, 자금 여유가 된다면 노도강 금관구 아파트를 사는것이 좋을것이라고 예상한다.
<br>
건축년도에 따른 집값 변화는 뚜렷하게 확인되지 않았고, 오히려 자치구가 집값에 더 영향을 주는것으로 해석되었다. <br>
<br>
마지막으로, 세대 당 인구수는 2002년 약2.9명에서, 2020년 약 2.1명으로 1인가구의 등장이 확대되고 있는것을 파악할 수 있었고, 자치구 중에서는 "관악구"가 2020년 기준 약 1.8명으로 자치구 중 세대당 인구수 최하위를 기록, 같은 기준 "양천구"는 2.51명으로 자치구중 세대당 인구수가 가장 많은 자치구로 해석되었다. <br>
이 결과는 곧 관악구의 A형 아파트에 대한 요구가 크다는 것을 의미하며, 시장가격은 곧 이를 반영할 것이므로, 정부의 적절하고 세밀한, 똑똑한 정책이 수반되어야 함을 의미한다. 반면 양천구 혹은 서초구, 노원구의 경우 세대당 인구수가 더 많은 상위권이므로 C형 이상 아파트에 살고 있는 주민이 많음을 의미한다. 따라서 향후 아파트 가격도 이러한 자치구의 특성을 반영할 것이라고 본다.<br>
