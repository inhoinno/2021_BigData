```python
dct2 = pd.Index(['TYPE', ])
data.rename(columns={data.columns[0]:dct2[0]}, inplace=True)
```

```python
data=data.set_index(['TYPE'])
```

```python
data
```

```python
data.index.size
```

```python
df0 = pd.DataFrame(data['2021. 04'].groupby(level=[0]).sum())
df0_sort = df0.groupby(level=['TYPE']).sum().sort_values(by='2021. 04', ascending=False)
df0_sort
```

```python
data = data.reindex(index=df0_sort.index)
data
```

```python
plt.rcParams["figure.figsize"] = (16,8)
plt.rcParams['lines.linewidth'] = 1

fig, ax = plt.subplots(figsize=(16,8))

for idx in range(0,data.index.size):
    plt.plot(data.columns.values, data.iloc[idx], marker='o')

plt.xticks(rotation=45)
plt.legend(data.index.values)
plt.title("2018-2021.04 업종별 KOSPI 지수",fontsize=24)
plt.show()
```
