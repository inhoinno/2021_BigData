```python
YYYY=2019
M=1
while(YYYY<2021):
    # File
    FILE = './1920_preprocessed/{year}_{month}_total.csv'.format(year=YYYY, month=M) 
    
    # Read CSV
    data = pd.read_csv(FILE,encoding='utf-8')
    
    data = data.drop(columns = data.columns[0])
    
    # 초기화
    data_part0 = data[0:18]
    data_part0 = data_part0.sort_index().reset_index(drop=True)
    colName = data_part0['corp'][0]
    data_part0 = data_part0.rename(columns={'2':colName})
    data_part0 = data_part0.drop(columns=data_part0.columns[0])
    result = data_part0
    
    # Data processing
    start = 18
    end = 36
    while(end <= 1818):
        data_part = data[start:end]
        data_part = data_part.sort_index().reset_index(drop=True)
        colName = data_part['corp'][0]
        data_part = data_part.rename(columns={'2':colName})
        data_part = data_part.drop(columns=data_part.columns[0])
        result = pd.merge(result, data_part)
    
        start += 18
        end += 18
    
    
    # 컬럼 처리
    result = result.rename(columns={'LPG충전소':'가스충전소'})
    result = result.rename(columns={'LPG 충전소':'가스충전소'})
    result = result.rename(columns={'pc방':'피시방'})
    result = result.rename(columns={'PC방':'피시방'})
    result = result.rename(columns={'교습소ㆍ공부방':'공부방'})
    result = result.rename(columns={'교습소･공부방':'공부방'})
    result = result.rename(columns={'교습소?공부방':'공부방'})
    result = result.rename(columns={'city':'A. 도시명'})
    result = result.rename(columns={'업종전체':'B. 업종전체'})
    result = result.rename(columns={' 업종전체':'B. 업종전체'})
    result = result.rename(columns={'업종':'B. 업종전체'})
    result = result.sort_index(axis=1)
    
    # Save to csv
    result.to_csv('./1920_processed/{year}_{month}_processed.csv'.format(year=YYYY,month=M),sep=',',encoding='utf-8') 
    
    
    print('{0}_{1}_fin'.format(YYYY,M))
    
    if(M + 1 >= 13): # M=12 Next year
        M = 1
        YYYY += 1
    else :     
        M += 1 
```
